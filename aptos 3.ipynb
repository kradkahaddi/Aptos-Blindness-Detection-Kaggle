```python
path_data = Path('/kaggle/input/aptos2019-blindness-detection/')
path_model = Path('/kaggle/input/aptos-pretrained-model-fast-ai/')
# path =      Path('/kaggle/input/aptos2019-blindness-detection')

path_train = path_data/'train_images'
path_test = path_data/'test_images'

train_df = pd.read_csv(path_data/'train.csv')
test_df = pd.read_csv(path_data/'test.csv')
```

```python
np.random.seed(17)
```

```python
train_df.head()
```

```python
data = ImageDataBunch.from_df(path_train, train_df, label_col =1, size = 128, test = path_test, suffix='.png').normalize(imagenet_stats)
learn = cnn_learner(data, models.resnet34, metrics = error_rate, pretrained =False)
```

```python
learn.load(path_model/'model_1');
```

```python
data.classes
```

```python
learn.fit_one_cycle(1)
```

```python
predictions = learn.get_preds(ds_type=DatasetType.Test)
```

```python
preds = np.argmax(np.array(predictions[0]), axis =1)
```

```python
np.sum(np.array(preds) ==4)/len(preds)
```
